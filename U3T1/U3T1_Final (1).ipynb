# Spacy

extração de entidades nomeadas (NER)


```python

import spacy
import pandas as pd
import networkx as nx
from spacy.lang.pt import Portuguese

# Carregar modelo spaCy para português
nlp = spacy.load("pt_core_news_sm")

# Função para extrair entidades nomeadas
def extract_entities(text):
    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        if ent.label_ in ["PER", "LOC", "ORG"]:  # Focar em pessoas, locais e organizações
            entities.append(ent.text)
    return entities

# Extrair entidades de ambos os textos
entities_1 = extract_entities(text_1)
entities_2 = extract_entities(text_2)

print("Entidades do Texto 1:", entities_1)
print("Entidades do Texto 2:", entities_2)
```

```python

# Função para criar um grafo baseado em entidades
def create_graph(entities):
    edges = []
    for i in range(len(entities)):
        for j in range(i + 1, len(entities)):
            edges.append((entities[i], entities[j]))
    return edges

# Criar grafos para os dois textos
edges_1 = create_graph(entities_1)
edges_2 = create_graph(entities_2)

G1 = nx.Graph()
G1.add_edges_from(edges_1)

G2 = nx.Graph()
G2.add_edges_from(edges_2)

# Visualizar métricas básicas
print("Grafo do Texto 1 - Número de nós:", G1.number_of_nodes(), "Número de arestas:", G1.number_of_edges())
print("Grafo do Texto 2 - Número de nós:", G2.number_of_nodes(), "Número de arestas:", G2.number_of_edges())
```

```python

import matplotlib.pyplot as plt

# Visualizar os grafos
plt.figure(figsize=(10, 6))
nx.draw(G1, with_labels=True, node_size=700, font_size=10)
plt.title("Grafo - Texto 1")
plt.show()

plt.figure(figsize=(10, 6))
nx.draw(G2, with_labels=True, node_size=700, font_size=10)
plt.title("Grafo - Texto 2")
plt.show()
```

```python

# Métricas de Rede
def analyze_graph_metrics(G, title):
    print(f"--- {title} ---")
    print(f"Número de Nós: {G.number_of_nodes()}")
    print(f"Número de Arestas: {G.number_of_edges()}")
    print(f"Densidade: {nx.density(G):.4f}")
    print(f"Centralidade de Grau (Top 5):")
    centrality = nx.degree_centrality(G)
    for node, cent in sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"  {node}: {cent:.4f}")
    print()

# Analisar métricas dos dois grafos
analyze_graph_metrics(G1, "Grafo do Texto 1")
analyze_graph_metrics(G2, "Grafo do Texto 2")
```

```python

from networkx.algorithms.community import louvain_communities

# Identificar comunidades usando Louvain
communities_1 = louvain_communities(G1)
communities_2 = louvain_communities(G2)

print("Número de Comunidades no Grafo do Texto 1:", len(communities_1))
print("Número de Comunidades no Grafo do Texto 2:", len(communities_2))

# Exibir as primeiras comunidades
print("Comunidades no Grafo do Texto 1:", communities_1[:3])
print("Comunidades no Grafo do Texto 2:", communities_2[:3])
```

# NLTK

Frequência de palavras e remoção de stopwords

```python
def analyze_word_frequencies(text, language='portuguese'):
    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words(language))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    freq_dist = FreqDist(filtered_tokens)
    return freq_dist.most_common(10)

# Analisar frequência de palavras nos textos
freq_text_1 = analyze_word_frequencies(text_1)
freq_text_2 = analyze_word_frequencies(text_2)

print("Frequência de Palavras - Texto 1:", freq_text_1)
print("Frequência de Palavras - Texto 2:", freq_text_2)
```

Analise sentimentos

```python
def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    return sia.polarity_scores(text)

# Analisar sentimentos dos textos
sentiment_text_1 = analyze_sentiment(text_1)
sentiment_text_2 = analyze_sentiment(text_2)

print("Sentimento - Texto 1:", sentiment_text_1)
print("Sentimento - Texto 2:", sentiment_text_2)
```

Complexidade textutal

```python
def analyze_text_complexity(text):
    sentences = nltk.sent_tokenize(text)
    tokens = word_tokenize(text)
    num_sentences = len(sentences)
    num_words = len(tokens)
    num_unique_words = len(set(tokens))
    avg_sentence_length = num_words / num_sentences
    lexical_density = num_unique_words / num_words
    return {
        "Número de Sentenças": num_sentences,
        "Número de Palavras": num_words,
        "Número de Palavras Únicas": num_unique_words,
        "Tamanho Médio das Sentenças": avg_sentence_length,
        "Densidade Lexical": lexical_density
    }

# Analisar complexidade dos textos
complexity_text_1 = analyze_text_complexity(text_1)
complexity_text_2 = analyze_text_complexity(text_2)

print("Complexidade Textual - Texto 1:", complexity_text_1)
print("Complexidade Textual - Texto 2:", complexity_text_2)
```

PoS Tagging

```python
def analyze_pos_tags(text):
    # Tokenizar o texto
    tokens = word_tokenize(text)

    # Filtrar apenas palavras alfanuméricas (removendo pontuações e símbolos)
    filtered_tokens = [word for word in tokens if word.isalnum()]

    # Aplicar PoS Tagging nos tokens filtrados
    pos_tags = pos_tag(filtered_tokens)

    # Contar a frequência das categorias gramaticais
    pos_counts = nltk.FreqDist(tag for word, tag in pos_tags)

    return pos_counts.most_common(5)

# Analisar PoS tags dos textos ajustados
pos_tags_text_1 = analyze_pos_tags(text_1)
pos_tags_text_2 = analyze_pos_tags(text_2)

print("PoS Tags - Texto 1:", pos_tags_text_1)
print("PoS Tags - Texto 2:", pos_tags_text_2)
```
