```python
"""
Questões cap 2
    2.1 a questão da as probabilidades de cara que na primeira linha da tabela esta o,5 e o,5 então para dar
    HH (cara e cara) tem que fazer 0,5 x 0,5 = 0,25. 
    HT (cara e coroa) para situação de dar coroa deve ser feito o valor menos 1,
    pois uma probabilidade tem que fechar em 1 sempre, se faz 1 - 0,5 = 0,5, (p1 = 0,5 x p2 = 0,5) = 0,25.
    TH (coroa e cara) mesma situação (1-0,5 = 0,5), 0,5 x 0,5 = 0,25. 
    TT(coroa e coroa) (p1 -1)x(p2-1) = 0,5.
    assim segue esse mesmo metodo para os seguintes valores da tabela.
"""

def calcular_resultados(p1_h, p2_h):
     # A probabilidade de obter coroa (T) é o complemento de obter cara (H), 1 - p1_h. e assim as seguites também.
    p1_t = 1 - p1_h 
    p2_t = 1 - p2_h 
    
    # Calculo das probabilidades dos resultados possíveis.
    hh = p1_h * p2_h        # H-H (cara-cara)
    ht = p1_h * p2_t        # H-T (cara-coroa)
    th = p1_t * p2_h        # T-H (coroa-cara)
    tt = p1_t * p2_t        # T-T (coroa-coroa)
    
    return hh, ht, th, tt   

probabilidades_tabela = [
    (0.5, 0.5),(0.6, 0.7),(0.4, 0.8),(0.1, 0.2),(0.3, 0.4)
]

# Calculando as probabilidades para cada par de p1(H) e p2(H)
resultados = []
for p1_h, p2_h in probabilidades_tabela:
    hh, ht, th, tt = calcular_resultados(p1_h, p2_h)
    resultados.append((p1_h, p2_h, hh, ht, th, tt))

# resultados
for resultado in resultados:
    p1_h, p2_h, hh, ht, th, tt = resultado
    print(f"Para p1(H)={p1_h} e p2(H)={p2_h}:")
    print(f" H-H: {hh:.2f}, H-T: {ht:.2f}, T-H: {th:.2f}, T-T: {tt:.2f}")
    print(" ")
```

```python
"""
Questão 2.2

    Fornecem um probelma com emails no total de 963 para ser resolvido com o teorema de bayes, P(a/b) = (P(b/a)xP(a))/P(b). 
    temos que 60% são spam então faz 1 - 0,6 = 0.4(40%) esse valor é de não spam, P(a) = 0,4. Sendo P(b) = 0,6
    ele fornece uma confiança de 98% (0.98), para resolver fazemos o seguinte para achar a insignificancia que é 1 - 0.98 = 0.02.
    então P(b/a) = 0,02. E agora resolvendo fica (0,02 x 0,4)/0,6 = 0,0133.
"""

P_a = 0.4  
P_b = 0.6 
P_b_a = 0.02  
Emails = 963
# Teorema de Bayes 
P_a_b = (P_b_a * P_a) / P_b
#Total
Total = P_a_b * Emails
print(f'A probabilidade de ser um e-mail não spam, dado que foi classificado como spam, é: {P_a_b:.5f}')
print(f'Emails não spam : {Total:.3f}')
```

```python
""" 
Questão 2.4
    
    criar um grafico da tabela fornecida e valores acumulativos, precisei usar da biblioteca para representar.

"""
    


import numpy as np
import matplotlib.pyplot as plt

# colocar os dados fornecidos em vetores
resultados = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilidades = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

# distribuição acumulada
probabilidade_acumulativa = np.cumsum(probabilidades)

#abrir grafico
fig, axs = plt.subplots(2, 1, figsize=(7, 5))

#função para massa de probabilidade
axs[0].bar(resultados, probabilidades, color='blue', alpha=0.7)
axs[0].set_title('Função Massa de Probabilidade (PMF)')
axs[0].set_xlabel('Resultados')
axs[0].set_ylabel('Probabilidade')

#funçao para grafico acumulada
axs[1].step(resultados,probabilidade_acumulativa , where='mid', color='red', alpha=0.7)
axs[1].set_title('Função de Distribuição Acumulada (CDF)')
axs[1].set_xlabel('Resultados')
axs[1].set_ylabel('Probabilidade Acumulada')

plt.tight_layout()
plt.show()
```

```python
"""
    Q3.3
    
"""
import matplotlib.pyplot as plt
import networkx as nx


g = nx.Graph()
g.add_nodes_from(['A','B','C','D','E','F','G','H'])
edges = [
    ('A', 'B', 1),  ('A', 'G', 2),     
    ('B', 'D', 2),  ('C', 'B', 2),   ('C', 'D', 1),  
    ('D', 'E', 1),  ('D', 'F', 1),  
    ('D', 'G', 2),  ('D', 'H', 1),  
     
]
for u, v, thickness in edges:
    g.add_edge(u, v, weight=thickness)

pos = nx.spring_layout(g, k=1.3)

edge_widths = [g[u][v]['weight'] for u, v, thickness in edges] 

nx.draw(g, pos, with_labels=True, node_color='blue', node_size=500, font_size=11, font_color='white', width=edge_widths)

for u, v, thickness in edges:
    mid_x = (pos[u][0] + pos[v][0]) / 2
    mid_y = (pos[u][1] + pos[v][1]) / 2
    plt.text(mid_x, mid_y, str(thickness), fontsize=20, ha='left', color='red')

plt.title("Grafo de Seguidores")
plt.axis('off')  
plt.show()
```
